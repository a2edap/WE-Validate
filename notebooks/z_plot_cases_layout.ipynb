```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# for Basemap to work in Python 3.8
import os
os.environ['PROJ_LIB'] = 'C:/Utilities/Python/Anaconda/Library/share'
from mpl_toolkits.basemap import Basemap

fino2_lat, fino2_lon = 55.006928, 13.154189
baltic2_lat, baltic2_lon = 54.9733, 13.1778

wfip2sodar_lat, wfip2sodar_lon = 45.57451, -120.74734

baltic2_df = pd.read_excel('../data/baltic2/baltic-2_layout.xlsx', sheet_name='edited')
```

```python
fig = plt.figure(num=None, figsize=(12, 8))
m = Basemap(projection='merc',
            llcrnrlat=54.5, urcrnrlat=55.5, llcrnrlon=11.8, urcrnrlon=14,
            resolution='h')
m.drawcoastlines()
m.fillcontinents(color='tan',lake_color='lightblue')
m.drawmapboundary(fill_color='lightblue')
# m.shadedrelief(scale=2)

x, y = m(fino2_lon, fino2_lat)
plt.plot(x, y, 'ob', markersize=2)

xt, yt = m(fino2_lon-0.18, fino2_lat+0.02)
plt.text(xt, yt, 'FINO2', c='b', fontsize=20)

for ind, row in baltic2_df.iterrows():
    x, y = m(row['lon'], row['lat'])
    plt.plot(x, y, '.r', markersize=1)
    
xt, yt = m(fino2_lon+0.07, fino2_lat-0.07)
plt.text(xt, yt, 'Baltic-2', c='r', fontsize=20)
    
plt.show()
```
